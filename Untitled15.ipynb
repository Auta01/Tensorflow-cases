```python
import tensorflow as tf
```

```python
#set random seed
tf.random.set_seed(42)

#1. create a model
model =tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

#2. compile the model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer= tf.keras.optimizers.SGD(),
              metrics =['accuracy'])
#3. fit the model
model.fit(x, y, epochs =5)
```

```python
# to visualize our model prediction lets create a function plot  decision boundary

import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(model, x, y):

  x_min, x_max= x[:, 0].min() -0.1, x[:,0].max() +0.1
  y_min, y_max= x[:, 1].min() -0.1, x[:,1].max() +0.1

  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                        np.linspace(y_min, y_max, 100))

  # create x value
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D array together

  #make predictions
  y_pred = model.predict(x_in)

  #check for multi-class
  if len(y_pred[0])>1:
    print('doing multiclass classification')

    # we have to reshape our prediction to get them ready for plotting
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
      print('doing binary classification')
      y_pred = np.round(y_pred).reshape(xx.shape)


  #plot the decision
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(x[:,0], x[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
#Check out the prediction our  model is making
plot_decision_boundary(model=model_3,
                       x=x,
                       y=y)
```

```python
import tensorflow as tf

# lets see if our model can be for regression problem...
tf.random.set_seed(42)

#create some regression data
x_regression = tf.range(0, 1000,5)
y_regression = tf.range(100,1100,5)#y = x+10

# split our regression data into training and test data
x_reg_train = x_regression[:150]
x_reg_test = x_regression[150:]
y_reg_train = y_regression[:150] # Corrected to use y_regression
y_reg_test = y_regression[150:]  # Corrected to use y_regression

# Reshape x_reg_train and x_reg_test to (None, 1) to be compatible with Dense layers
x_reg_train = tf.expand_dims(x_reg_train, axis=-1)
x_reg_test = tf.expand_dims(x_reg_test, axis=-1)
```

Oh wait..... we compiled our model for a binary classiffication problem but we are working on a regression problem lets change the model to suit the data

```python
#set a random seed
tf.random.set_seed(42)
#create the model
model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#compile the model
model_3.compile(loss= tf.keras.losses.mae,
                optimizer =tf.keras.optimizers.Adam(),
                metrics=['mae'])

#fit the model
model_3.fit(x_reg_train, y_reg_train, epochs=100)
```

```python
#make prediction with our train data
y_reg_preds= model_3.predict(x_reg_test)
```

```python
#plot the model predictions against regression data
plt.figure(figsize=(10,7))
plt.scatter(x_reg_train, x_reg_train, c ='b', label='Training Data' )
plt.scatter(x_reg_test, x_reg_test, c='g', label='Test Data' )
plt.scatter(x_reg_test, y_reg_preds, c= 'r',label= 'Prediction')
```
